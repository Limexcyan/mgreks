```python
# Dowolny zbiór liczb naturalnych
s = random_integer_set(1, 10, 6)
print(f"Random set: {s}")

all_partitions = []

# Podziały zbioru s
for i, p in enumerate(partitions(s), 1):
    all_partitions.append(p)
    print(f"{i}: {p}")

# Zbiór wszystkich podziałów zbioru s, to jest D(S).
print(f"All partitions: {all_partitions}")
```

```python
# Sprawdź, czy podział jest niekrzyżujący się
rand_part = random_partition(s)
print(rand_part)
print(is_noncrossing(rand_part))
```

```python
# Dla zbioru s podziel D(s) na NC(s) i pozostałe podziały
noncrossing = []
crossing = []

for p in partitions(s):
    if is_noncrossing(p):
        noncrossing.append(p)
    else:
        crossing.append(p)

print(f'Non-crossing partitions of a set {s}:')
print(noncrossing)

print(f'Crossing partitions of a set {s}')
print(crossing)

# lub użyj
print(non_crossing_partitions(s))
```

```python
def benchmark_drawing_times(max_n=6):
    draw_times = []
    sizes = []

    for n in range(1, max_n + 1):
        s = list(range(1, n + 1))
        print(f"Generating for set: {s}")

        all_parts = list(partitions(s))
        nc_parts = [p for p in all_parts if is_noncrossing(p)]

        nodes, edges = build_hasse_edges(nc_parts)

        print(f"Partitions: {len(all_parts)} total, {len(nc_parts)} non-crossing")

        start = time.perf_counter()
        draw_hasse_lattice_manual(nodes, edges, f"Non-Crossing Lattice for n={n}")
        end = time.perf_counter()
        elapsed = end - start
        draw_times.append(elapsed)
        sizes.append(n)

        print(f"Draw time for n={n}: {elapsed:.4f} seconds")

    return sizes, draw_times

def plot_drawing_benchmark(sizes, draw_times, filename="drawing_times", size=0):
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, draw_times, marker='o', color='darkred')
    plt.xlabel("n (size of set [1..n])")
    plt.ylabel("Time to draw lattice (seconds)")
    plt.title("Time to Draw Non-Crossing Partition Lattice")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{filename}{size}.png")
    plt.show()

max_n = 6
sizes, draw_times = benchmark_drawing_times(max_n=max_n)
plot_drawing_benchmark(sizes, draw_times, size=max_n)
```
